# Problem 1 Part A: [p<sub>l</sub> = 1/3], [p<sub>r</sub> = 1/3], and [p<sub>s</sub> = 1/3]

```python
import numpy as np

def random_walker(t):
    """ Runs random walk for t time, with each t equal to one step.
    pl = 1/3, pr = 1/3, and ps = 1/3.
    t argument = time for random walker to run."""

    x = 0     # current position
    
    for i in range(t):
        iteration_choice = np.random.random()

        if iteration_choice <= 0.333333333:
            x += -1 # move left
        elif iteration_choice >= 0.333333333 and iteration_choice <= 0.666666666:
            x += 1 # move right
        elif iteration_choice >= 0.666666666:
            pass #stay
    
    return x

def many_walks(n, t):
    """returns a list of many walks for the specified time
    t argument = time for random walker to run.
    n argument = number of walks.
    """
    many_walks = []

    for i in range(n):
        many_walks.append(random_walker(t))

    return many_walks

#warning will take about >15 seconds due to 40k walks
t10_many_walks = many_walks(40000, 10)
t100_many_walks = many_walks(40000, 100)
t1000_many_walks = many_walks(40000, 1000)

```

```python
import matplotlib.pyplot as plt

#plots for t=10/100/1000 
plt.hist(t10_many_walks, alpha=0.55, density=True, label='t=10', bins = range(min(t10_many_walks),max(t10_many_walks),))
plt.hist(t100_many_walks, alpha=0.55, density=True, label='t=100', bins = range(min(t100_many_walks),max(t100_many_walks), 3))
plt.hist(t1000_many_walks, alpha=0.55, density=True, label='t=1000', bins = range(min(t1000_many_walks),max(t1000_many_walks), 3))
#plot specifications
plt.legend(loc='upper right')
plt.xlabel('Walk End Position', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('1D Random Walker of Equal Movement Probability', fontweight='bold', fontsize = 16)
plt.xticks(range(-120,120,20))
plt.figure(figsize=(8,4), dpi= 300)
```

```python
import numpy as np

#np.hist() returns a hist array and an array of the bin edges
hist10_raw, bin_edges10 = np.histogram(t10_many_walks, bins = range(min(t1000_many_walks), max(t1000_many_walks)),  density = True) #min and max bin changed to extend tails to match t=1000

hist10 = np.append(hist10_raw, 0) #because plot must have same dimensions; the histogram has one less due to bin edges being n-1

#makes a cumulative array of the previous histogram
cum_hist10 = np.cumsum(hist10)


#repeat above for t=100 and t=1000
hist100_raw, bin_edges100 = np.histogram(t100_many_walks, bins = range(min(t1000_many_walks), max(t1000_many_walks)),  density = True) #min and max bin changed to extend tails to match t=1000
hist100 = np.append(hist100_raw, 0) #because plot must have same dimensions; the histogram has one less due to bin edges being n-1
cum_hist100 = np.cumsum(hist100)

hist1000_raw, bin_edges1000 = np.histogram(t1000_many_walks, bins = range(min(t1000_many_walks), max(t1000_many_walks)),  density = True)
hist1000 = np.append(hist1000_raw, 0) #because plot must have same dimensions; the histogram has one less due to bin edges being n-1
cum_hist1000 = np.cumsum(hist1000)

#plot the cumulative hist array and bin edges for t=10/100/1000
plt.plot(bin_edges10, cum_hist10, label='t=10',)
plt.plot(bin_edges100, cum_hist100, label='t=100',)
plt.plot(bin_edges1000, cum_hist1000, label='t=1000',)
#plot specifications
plt.xticks(range(-120,120,20))
plt.legend(loc='upper left')
plt.xlabel('Walk End Position', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('1D Random Walker of Equal Movement Probability', fontweight='bold', fontsize = 16)
plt.figure(figsize=(8,4), dpi= 300)


plt.show()
```

# Part B: [p<sub>l</sub> = 0], [p<sub>r</sub> = 1/2], and [p<sub>s</sub> = 1/2]

```python
import numpy as np

def random_walker_biased(t):
    """ Runs random walk for t time, with each t equal to one step.
    pl = 0, pr = 1/2, and ps = 1/2
    t argument = time for random walker to run."""

    x = 0     # current position
    
    for i in range(t):
        iteration_choice = np.random.random()

        if iteration_choice <= 0.5:
            x += 1 # move right
        elif iteration_choice > 0.5:
            pass # stay
    
    return x

def many_walks_biased(n, t):
    """returns a list of many walks for the specified time
    t argument = time for random walker to run.
    n argument = number of walks.
    """
    many_walks = []

    for i in range(n):
        many_walks.append(random_walker_biased(t))

    return many_walks

#warning will take about >10 seconds due to 40k walks
t10_many_walks_biased = many_walks_biased(40000, 10)
t100_many_walks_biased = many_walks_biased(40000, 100)
t1000_many_walks_biased = many_walks_biased(40000, 1000)
```

```python
import matplotlib.pyplot as plt

#plots for t=10/100/1000 
plt.hist(t10_many_walks_biased, alpha=0.55, density=True, label='t=10', bins = range(min(t10_many_walks_biased),max(t10_many_walks_biased),))
plt.hist(t100_many_walks_biased, alpha=0.55, density=True, label='t=100', bins = range(min(t100_many_walks_biased),max(t100_many_walks_biased), 3))
plt.hist(t1000_many_walks_biased, alpha=0.55, density=True, label='t=1000', bins = range(min(t1000_many_walks_biased),max(t1000_many_walks_biased), 3))
#plot specifications
plt.legend(loc='upper right')
plt.xlabel('Walk End Position', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('1D Random Walker in Absence of Left Movement Probability', fontweight='bold', fontsize = 16)
plt.xticks(range(0,600,40))
plt.figure(figsize=(8,4), dpi= 300)
```

```python
import numpy as np

#np.hist() returns a hist array and an array of the bin edges
hist10_raw, bin_edges10 = np.histogram(t10_many_walks_biased, bins = range(0, max(t1000_many_walks_biased)),  density = True) #max bin changed to extend tail to match t=1000

hist10 = np.append(hist10_raw, 0) #because plot must have same dimensions; the histogram has one less due to bin edges being n-1

#makes a cumulative array of the previous histogram
cum_hist10 = np.cumsum(hist10)


#repeat above for t=100 and t=1000
hist100_raw, bin_edges100 = np.histogram(t100_many_walks_biased, bins = range(0, max(t1000_many_walks_biased)),  density = True) #min and max bin changed to extend tail to match t=1000
hist100 = np.append(hist100_raw, 0) #because plot must have same dimensions; the histogram has one less due to bin edges being n-1
cum_hist100 = np.cumsum(hist100)

hist1000_raw, bin_edges1000 = np.histogram(t1000_many_walks_biased, bins = range(0, max(t1000_many_walks_biased)),  density = True) #min  bin changed to extend tail to match t=10/100
hist1000 = np.append(hist1000_raw, 0) #because plot must have same dimensions; the histogram has one less due to bin edges being n-1
cum_hist1000 = np.cumsum(hist1000)

#plot the cumulative hist array and bin edges for t=10/100/1000
plt.plot(bin_edges10, cum_hist10, label='t=10',)
plt.plot(bin_edges100, cum_hist100, label='t=100',)
plt.plot(bin_edges1000, cum_hist1000, label='t=1000',)
plt.xticks(range(0,max(t1000_many_walks_biased),50))
plt.legend(bbox_to_anchor=(1, 1))
plt.xlabel('Walk End Position', fontweight='bold')
plt.ylabel('Probability', fontweight='bold')
plt.title('1D Random Walker in Absence of Left Movement Probability', fontweight='bold', fontsize = 16)
plt.figure(figsize=(8,4), dpi= 300)


plt.show()
```
